```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-3, 3, 100)
y = relu(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label="ReLU", color='blue')
plt.title("Fonction d'activation : ReLU")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.savefig("relu_activation.png", dpi=300)
plt.show()
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Stable softmax
    return exp_x / np.sum(exp_x)

x = np.linspace(-3, 3, 100)
y = softmax(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label="Softmax", color='orange')
plt.title("Fonction d'activation : Softmax")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.savefig("softmax_activation.png", dpi=300)
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

x = np.linspace(-3, 3, 100)
y = tanh(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label="Tanh", color='green')
plt.title("Fonction d'activation : Tanh")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.savefig("tanh_activation.png", dpi=300)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-3, 3, 100)
y = sigmoid(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label="Sigmoïde", color='red')
plt.title("Fonction d'activation : Sigmoïde")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.savefig("sigmoïde_activation.png", dpi=300)
plt.show()
```

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

x = np.linspace(-3, 3, 100)
y = leaky_relu(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label="Leaky ReLU", color='purple')
plt.title("Fonction d'activation : Leaky ReLU")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.savefig("leaky_relu_activation.png", dpi=300)
plt.show()
```

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

x = np.linspace(-3, 3, 100)
y = elu(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label="ELU", color='brown')
plt.title("Fonction d'activation : ELU")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.savefig("elu_activation.png", dpi=300)
plt.show()
```

```python
def mse(theta):
    return (theta - 2)**2 + 3

theta_vals = np.linspace(-2, 6, 500)
mse_vals = mse(theta_vals)

theta_start = 5
mse_start = mse(theta_start)

theta_min = 2
mse_min = mse(theta_min)

plt.figure(figsize=(8, 6))
plt.plot(theta_vals, mse_vals, label="Fonction de coût MSE", color="blue")
plt.scatter(theta_start, mse_start, color="red", label="Point de départ ($\\theta_{start}$)", zorder=5)
plt.scatter(theta_min, mse_min, color="green", label="Minimum ($\\theta_{min}$)", zorder=5)

plt.annotate('Point de départ',
             xy=(theta_start, mse_start), xytext=(theta_start - 1, mse_start + 5),
             arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=10)

plt.annotate('Objectif (Minimum)',
             xy=(theta_min, mse_min), xytext=(theta_min + 1, mse_min + 5),
             arrowprops=dict(facecolor='green', arrowstyle='->'), fontsize=10)

plt.title("Exemple de Fonction de Coût MSE et Optimisation", fontsize=14)
plt.xlabel("$\\theta$", fontsize=12)
plt.ylabel("MSE($\\theta$)", fontsize=12)
plt.axhline(y=mse_min, color="gray", linestyle="--", alpha=0.6, label="Niveau du minimum")
plt.legend()
plt.grid(alpha=0.3)
plt.savefig("cost_function.png", dpi=300)
plt.show()
```

```python
def f(x):
    return x**2

def df(x):
    return 2 * x

# Parameters for three scenarios
learning_rates = [1.009, 0.01, 0.1]  # Too high, too low, perfect
titles = ["Learning Rate Too High", "Learning Rate Too Low", "Perfect Learning Rate"]
x_start = 4
iterations = 10

# Plot for each scenario
for i, learning_rate in enumerate(learning_rates):
    # Gradient descent
    x_vals = [x_start]
    costs = [f(x_start)]
    for _ in range(iterations):
        grad = df(x_vals[-1])
        x_next = x_vals[-1] - learning_rate * grad
        x_vals.append(x_next)
        costs.append(f(x_next))
    
    # Plot
    x = np.linspace(-5, 5, 500)
    plt.figure(figsize=(8, 6))
    
    # Function plot
    plt.plot(x, f(x), label="Function $f(x) = x^2$", color="blue")
    
    # Gradient descent steps on cost function
    plt.scatter(x_vals, costs, color="red", label="Gradient Descent Steps")
    for j in range(len(x_vals) - 1):
        plt.arrow(x_vals[j], costs[j], 
                  x_vals[j+1] - x_vals[j], costs[j+1] - costs[j], 
                  head_width=0.2, head_length=0.2, color="orange", length_includes_head=True)
    
    # Labels and titles
    plt.title(titles[i])
    plt.xlabel("$x$")
    plt.ylabel("$f(x)$")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.savefig(f'learning_rate{i}')
    plt.show()
```

```python
def cost_function(x):
    return np.sin(3 * x) + 0.5 * x**2

x_vals = np.linspace(-2, 2, 500)
y_vals = cost_function(x_vals)

np.random.seed(42)
steps = [-1.8]
for _ in range(10):
    grad = 3 * np.cos(3 * steps[-1]) + steps[-1]  # Gradient of the cost function
    noise = np.random.normal(0, 0.1)  # Add stochastic noise
    step = steps[-1] - 0.2 * grad + noise  # Gradient descent step with noise
    steps.append(step)

step_costs = [cost_function(step) for step in steps]

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="Cost Function $f(x)$", color="blue")

plt.scatter(steps, step_costs, color="red", label="SGD Steps")
for i in range(len(steps) - 1):
    plt.arrow(steps[i], step_costs[i],
              steps[i + 1] - steps[i], step_costs[i + 1] - step_costs[i],
              head_width=0.05, head_length=0.1, color="orange", length_includes_head=True)

# Add labels and title
plt.title("Descente de Gradient Stochastique sur Fonction Non-Convexe")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend()
plt.grid(alpha=0.3)
plt.savefig("stochastic_gradient_descent.png")
plt.show()
```
