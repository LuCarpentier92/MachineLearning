```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Data/exemple_simple_kmeans-3.csv', sep=',' )
plt.figure(1, figsize=(5, 5)) 
plt.scatter(data['X'], data['Y'],s=40)
```

```python
fig, axs = plt.subplots(4, 3) #fig contient tous les sous graphiques et axs est un array de sous graphe
inertie_lst = []
k=1
for row in range(4):
    for col in range(3):
        print("k= ",k)
        model = KMeans(n_clusters=k)
        model.fit(data)    
        inertie_lst.append(model.inertia_)
        centers = model.cluster_centers_
        axs[row,col].set_title(f"k = {str(k)}, inertie = {round(model.inertia_,2)}")
        axs[row,col].scatter(centers[:,0], centers[:,1],c='red',s=100)
        axs[row,col].scatter(data.X, data.Y,c='grey',s=1)
        for i in range(len(model.labels_)) :
            axs[row,col].text(data.X[i], data.Y[i], model.labels_[i])
        k+=1
        
fig.set_figheight(20)
fig.set_figwidth(15)
              
```

```python
fig, ax = plt.subplots(1)
ax.set_xticks(np.arange(15))  #met des marque de 0 à 14
ax.plot(range(1,13), inertie_lst)  # k de 1 à 12 et les valeurs de la liste
ax.set_xlabel("k", fontsize=20) 
ax.set_ylabel("Inertie", fontsize=20)
fig.set_figheight(8)#inches
fig.set_figwidth(8) 
plt.grid()
""" On l'appelle la méthode Elbow (méthode du coude) car il faut interpreter la position k = 1 comme l'épaule et k = 12 comme la main. Entre l"épaule et la main on cherche le coude, soit l'endroit pour lequel la diminution de l'inertie devient quasi nulle. Ici nous trouvons k = 9. Bien que ce résultat pouvait sembler évident, il faut tout de même rappeler que nous travaillons en 2D ce qui rend la visualisation simple. En revanche, avec 36 dimensions, bon courage pour déterminer cette valeur à loeil :)"""
```
