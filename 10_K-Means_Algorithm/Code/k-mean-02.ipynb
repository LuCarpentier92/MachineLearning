```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Data/exemple_simple_kmeans-2.csv', sep=',' )
plt.figure(1, figsize=(10, 10))
plt.scatter(data.X, data.Y,s=60)
plt.ylim(0,6) 
plt.xlim(0,15)
```

```python
model = KMeans(n_clusters=2)
model.fit(data)

        
colormap = np.array(['green','blue'])
plt.figure( figsize=(10, 10))
plt.scatter(data['X'], data['Y'],c=colormap[model.labels_],s=2)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1],c='red',s=500)
for i in range(len(model.labels_)) :
    plt.text(data.X[i], data.Y[i],model.labels_[i], fontsize=10 ) 
plt.ylim(0,6) 
plt.xlim(0,15)
```

```python
#model.labels_
model.inertia_/len(data.X)
```
